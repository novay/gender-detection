```python
import pandas as pd 
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
df = pd.read_csv('./data/nama-gender-split.csv')
df = df.fillna('')
df.tail()
```

```python
def generate_wordcloud(column):
    text = ' '.join(column.dropna().values.astype(str))
    wordcloud = WordCloud(width = 1200, 
                          height = 800, 
                          colormap='Set2', 
                          background_color='white', 
                          random_state=2, 
                          collocations=False).generate(text) 

    return wordcloud.to_image()
```

```python
fig, axs = plt.subplots(1, 3, figsize=(25, 25))

axs[0].imshow(generate_wordcloud(df[df['gender'] == 'm']['nama_awal']))
axs[0].set_title('First name [M]')
axs[0].axis('off')

axs[1].imshow(generate_wordcloud(df[df['gender'] == 'm']['nama_tengah']))
axs[1].set_title('Middle name [M]')
axs[1].axis('off')

axs[2].imshow(generate_wordcloud(df[df['gender'] == 'm']['nama_akhir']))
axs[2].set_title('Last name [M]')
axs[2].axis('off')

fig.show()
fig.savefig('./assets/male-name.png', bbox_inches='tight')
```

```python
fig, axs = plt.subplots(1, 3, figsize=(25, 25))

axs[0].imshow(generate_wordcloud(df[df['gender'] == 'f']['nama_awal']))
axs[0].set_title('First name [F]')
axs[0].axis('off')

axs[1].imshow(generate_wordcloud(df[df['gender'] == 'f']['nama_tengah']))
axs[1].set_title('Middle name [F]')
axs[1].axis('off')

axs[2].imshow(generate_wordcloud(df[df['gender'] == 'f']['nama_akhir']))
axs[2].set_title('Last name [F]')
axs[2].axis('off')

fig.show()
fig.savefig('./assets/female-name.png', bbox_inches='tight')
```
