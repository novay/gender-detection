```python
import pandas as pd

df = pd.read_csv('./data/nama-gender.csv')

print("Total Data: "+str(len(df)))
df.sample(5)
```

```python
df.to_csv('./data/nama-gender-titik.csv', sep=';', index=False)
```

```python
df[['nama_awal', 'nama_tengah', 'nama_akhir']] = df['nama'].str.split(n=2, expand=True);
df = df.fillna('')
df.to_csv('./data/nama-gender-split.csv')
df.tail()
```

```python
value_counts = df['gender'].value_counts(dropna = False)
print(value_counts)

total_data = df['gender'].count()
jumlah_m = df['gender'].value_counts()['m']
jumlah_f = df['gender'].value_counts()['f']

persentase_m = (jumlah_m / total_data) * 100
persentase_f = (jumlah_f / total_data) * 100

print()
print("Persentase Male: {:.2f}%".format(persentase_m))
print("Persentase Female: {:.2f}%".format(persentase_f))
```

```python
alphabets= ['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z']
total_percent=0
total_count = 0
startletter_count = {}
for i in alphabets:
    print('Number of names starting with {} are '.format(i),len(df[df['nama'].str.startswith(i)]))
    startletter_count[i] = len(df[df['nama'].str.startswith(i)])
    print('Percent: ',len(df[df['nama'].str.contains(i)])/len(df['nama'])*100)
    total_count = total_count + len(df[df['nama'].str.contains(i)])
    total_percent=total_percent+(len(df[df['nama'].str.contains(i)])/len(df['nama'])*100)
print('Total names = ', total_count)
print("Total percentage = ", total_percent)
```

```python
print(startletter_count)
```

```python
plt.figure(figsize = (16,8))
plt.bar(startletter_count.keys(),startletter_count.values())
plt.xlabel('Starting alphabet')
plt.ylabel('No. of names')
plt.title('Number of names starting with each letter')
```

Top 5 alphabets with which most of the names start are:

```python
print('The 5 most name starting letters are : ', *sorted(startletter_count.items(), key=lambda item: item[1])[-5:][::-1])
```

How many names end with the alphabets a,b...z

```python
small_alphabets = ['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','x','y','z']
endletter_count ={}
for i in small_alphabets:
    endletter_count[i]=len(df[df['nama'].str.endswith(i)])
print(endletter_count)
```

```python
plt.figure(figsize = (16,8))
plt.bar(endletter_count.keys(),endletter_count.values())
plt.xlabel('Ending alphabet')
plt.ylabel('No. of names')
plt.title('Number of names ending with each letter')
```

```python
print('The 5 most name ending letters are : ', *sorted(endletter_count.items(), key=lambda item: item[1])[-5:][::-1])
```

```python
from wordcloud import WordCloud, STOPWORDS

text =  " ".join(i for i in df.nama)
word_cloud = WordCloud(
        width=3000,
        height=2000,
        random_state=1,
        background_color="white",
        colormap="BuPu",
        collocations=False,
        stopwords=STOPWORDS,
        ).generate(text)

plt.imshow(word_cloud)
plt.axis("off")    
plt.show()
```
